```julia
using Plots

# Parameters
R = 10.0           # Resistance in ohms
L = 1e-3          # Inductance in henries
C = 1e-6           # Capacitance in farads
V0 = 5.0           # Peak voltage in volts
ω_range = range(0, 6e4, length=3000)  # Angular frequency range (rad/s)

# Time range for the sine wave (one full period of the lowest frequency)
t = range(0, 0.1, length=1000)  # Time from 0 to 1ms

# Impedance functions
Z_RL = R .+ 1im .* ω_range .* L
Z_C = 1.0 ./ (1im .* ω_range .* C)

# Total impedance
Z_total = 1.0 ./ (1.0 ./ Z_RL .+ 1.0 ./ Z_C)

# Current as a function of angular frequency
I_amplitude = V0 ./ abs.(Z_total)

# Calculate the voltage as a sine wave in time
V_in_time = V0 .* sin.(ω_range[1] .* t)  # Voltage for one frequency (e.g., ω_range[1])

# Resonance frequency
ω_res = sqrt((1/(L * C) - (R^2/L^2)))
f_res = ω_res / (2 * π)

# Plot current amplitude vs angular frequency
plot(ω_range, I_amplitude, xlabel="Angular Frequency (rad/s)", ylabel="Current Amplitude (A)",
     label="Current Amplitude", lw=2, legend=:topright)
vline!([ω_res], label="Resonance Frequency", lw=2, color=:red)
ylims!(0.0,1.0)
title!("Current Amplitude vs Angular Frequency")

# Plot voltage as a sine function of time
plot(t, V_in_time, xlabel="Time (s)", ylabel="Voltage (V)", label="Voltage (ω₁ = $(ω_range[1]) rad/s)", lw=2)
title!("Voltage vs Time for ω₁ = $(ω_range[1]) rad/s")

println("Resonance angular frequency: $ω_res rad/s")
println("Resonance frequency: $f_res Hz")
```
